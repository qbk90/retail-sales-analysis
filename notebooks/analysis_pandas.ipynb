```python
import pandas as pd

# Ruta del archivo CSV que contiene los datos a ser analizados
ruta_archivo = '../data/retail_sales_dataset.csv'

df = pd.read_csv(ruta_archivo)

df
```

```python
# Muestra las primeras 10 filas del DataFrame

df.head(10)
```

```python
# Muestra las ultimas 5 filas del DataFrame

df.tail(5)
```

```python
# Metodo info() para obtener informacion general del DataFrame

df.info()
```

```python
# Generar estadisticas descriptivas con el metodo describe()

df.describe()
```

```python
# Inspecciona los tipos de datos usando el atributo dtypes

df.dtypes
```

```python
# Valores unicos de la columna Product Category

df['Product Category'].value_counts()
```

```python
# Valores de la columna Gender

df['Gender'].unique()
```

```python
# Mostrar filas con ventas mayores a 3

condicion = df['Quantity'] > 3

df[condicion]
```

```python
# Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.

condicion = df['Price per Unit'] > 50

df[condicion]
```

```python
# Utilizando el método query(),
# filtra el DataFrame para mostrar las filas donde el Product Category sea Electronics
# y las ventas sean mayores a 2.

df.query("(`Product Category` == 'Electronics') & (Quantity > 2)")
```

```python
# Selecciona y muestra solo las columnas Product Category y Total Amount del DataFrame.

columnas = ['Product Category', 'Total Amount']

df[columnas]
```

```python
# Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive)
# y las columnas Product Category y Customer ID.

df.loc[5:10, ['Product Category', 'Customer ID']]
```

```python
# Utilizando iloc[], selecciona y muestra las primeras 5 filas y
# las primeras 3 columnas del DataFrame.

df.iloc[:5, :3]
```
